```python
huelgas_df = pd.read_csv('csv/huelgas.csv', sep='\t')

salarios_df = pd.read_csv('csv/sal_med.csv', sep='\t')
print(salarios_df.info())

comunidades_interes = ['Andalucía', 'Aragón', 'Asturias, Ppado de', 'Balears,Illes', 'Canarias', 'Cantabria', 'Castilla y León',
                       'Castilla-La mancha', 'Madrid, Comunidad de', 'Cataluña', 'Comunidad Valenciana', 'Extremadura', 'Galicia', 'Murcia, Región de',
                      'Navarra, Comunidad Foral', 'País Vasco', 'Rioja, La']
huelgas_df = huelgas_df[huelgas_df['Comunidad Autónoma y Provincia'].isin(comunidades_interes)]

datos_huelgas_desarrolladas = huelgas_df[huelgas_df['Tipo de indicador'] == 'HUELGAS DESARROLLADAS']
print("dhd: ", datos_huelgas_desarrolladas.info())
total_huelgas_desarrolladas = huelgas_df.loc[huelgas_df['Tipo de indicador'] == 'HUELGAS DESARROLLADAS', 'Total']
print("resultado: ", total_huelgas_desarrolladas)

print("datos1: ", huelgas_df[huelgas_df['Comunidad Autónoma y Provincia'].isin(comunidades_interes)]['Tipo de indicador'].unique())
print("datos2: ", huelgas_df[huelgas_df['Tipo de indicador'] == 'HUELGAS DESARROLLADAS'])
print("Cantidad de filas en huelgas_df:", huelgas_df.shape[0])
print("Cantidad de filas en salarios_df:", salarios_df.shape[0])

huelgas_df['Total'] = huelgas_df['Total'].str.replace(',', '')
huelgas_df['Total'] = huelgas_df['Total'].str.replace('.', '')
huelgas_df = huelgas_df[huelgas_df['Total'] != '']
huelgas_df['Total'] = huelgas_df['Total'].astype(float)
salarios_df['Total'] = salarios_df['Total'].str.replace(',', '')
salarios_df['Total'] = salarios_df['Total'].str.replace('.', '')
salarios_df = salarios_df[salarios_df['Total'] != '']
salarios_df['Total'] = salarios_df['Total'].astype(float)



X = huelgas_df['Total'].values.reshape(-1, 1)  # Cantidad de huelgas como variable independiente
y = salarios_df['Total'].values  # Salario medio como variable dependiente

modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.xlabel('Cantidad de Huelgas')
plt.ylabel('Salario Medio')
plt.title('Relación entre la Cantidad de Huelgas y el Salario Medio')
plt.legend()
plt.show()
```
