```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('csv/Evolution_DataSets.csv')

columns = ['Specie', 'Cranial_Capacity']
drop_duplicates = df.drop_duplicates('Specie')
max_cranial = drop_duplicates.sort_values('Cranial_Capacity', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(max_cranial['Specie'], max_cranial['Cranial_Capacity'], color='skyblue')

plt.xlabel('Especie')
plt.ylabel('Capacidad Craneal')
plt.title('Especies de homnidos con la mayor capacidad craneal')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
