```python
import matplotlib.pyplot as plt
import pandas as pd


# Leer el archivo CSV
df_evo = pd.read_csv('csv/Evolution_DataSets.csv')

# Seleccionar las especies de interés
especies_de_interes = ['Homo Neanderthalensis\n', 'Homo Erectus', 'Homo Sapiens', 'Homo Habilis', 'Homo Heidelbergensis']
df_especies_de_interes = df_evo[df_evo['Specie'].isin(especies_de_interes)]

# Graficar un scatter plot con cada especie coloreada
plt.figure(figsize=(10, 6))

for especie in especies_de_interes:
    subset = df_evo[df_evo['Specie'] == especie]
    plt.scatter(subset['Cranial_Capacity'], subset['Height'], label=especie, edgecolors='k')

# Añadir etiquetas y título
plt.xlabel('Capacidad Craneal')
plt.ylabel('Altura')
plt.title('Clasificación de especies')
plt.legend()
plt.grid(True)
plt.show()
```
